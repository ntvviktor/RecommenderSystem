```python
def get_five_reviews(group):
    return group.head(5)
new_df = reviews.groupby('isbn', group_keys=False).apply(get_five_reviews)
```

```python
new_df.shape
```

```python
list_of_reviews = new_df["review_sentences"].to_list()
reviews_list = []
for reviews in list_of_reviews:
    review_dict = {}
    for each_sent in reviews:
        if each_sent[0] not in review_dict:
            review_dict[each_sent[0]] = each_sent[1]
        else:
            review_dict[each_sent[0]] += " " + each_sent[1]
    
    if len(review_dict) > 1:
        para = ""
        for v in review_dict.values():
            para +=  "\n" + v
        reviews_list.append(para)
    else: 
        for v in review_dict.values():
            reviews_list.append(v)

len(reviews_list)
```

```python
new_df["reviews"] = reviews_list
```
